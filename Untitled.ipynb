## Plotting Lesson

```python
import numpy
import os
```

```python
distance_file = os.path.join('data','distance_data_headers.csv')
```

```python
distances = numpy.genfromtxt(fname=distance_file, delimiter=',',dtype='unicode')
```

```python
print(distances)
```

```python
header= distances[0]
data = distances[1:]
data = data.astype(float)
```

```python
print(header)
print(data)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.plot(data[:,0],data[:,1])
```

```python

sample = header[1]

plt.figure()
for col in range(1, len(header)):
    fig = plt.plot(data[:,0],data[:,col], label=header[col])
plt.xlabel('simulation Frame')
plt.ylabel('Distance (angstroms)')

plt.legend()

plt.savefig('all_samples_300dpi.png', dpi=300)
```

```python
for col in range(1, len(header)):
    plt.figure()
    fig = plt.plot(data[:,0],data[:,col], label=header[col])
    plt.xlabel('simulation Frame')
    plt.ylabel('Distance (angstroms)')
    plt.legend()
    plt.savefig(F'{header[col]}.png', dpi=300)
```
